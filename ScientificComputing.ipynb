## Gauss Seidel

```python
import numpy as np

def gaussSeidel(x, y, eps = 0.01, t = 15):
  x = np.array(x)
  y = np.array(y)
  
  #diagonally dominant?
  diag = np.diag(np.abs(x))
  np.fill_diagonal(x, 0)
  sumX = np.sum(np.abs(x), axis = 1)
  
  if not np.all(diag > sumX) :
    print("NOT DIAGONALLY DOMINANT")
    return False
  else:
    print("DIAGONALLY DOMINANT")
  
  x = -x
  mat_old = np.zeros(x[0].shape)

  for i in range(t):
    mat_new = np.array(mat_old)

    for j, num in enumerate(x):
      mat_new[j] = (y[j] + np.dot(num, mat_new))/ diag[j]
    
    print("iter - ", i, mat_new)

    selisih = np.sqrt(np.dot(mat_new - mat_old, mat_new - mat_old))
    if selisih <= eps:
      return True
    
    mat_old = mat_new
  
  return False
```

```python
x = np.array([10, 4, 5, 1, 6, 1, 3, 1, 6]).reshape(3, 3)
y = np.array([8, 5, 10])

if gaussSeidel(x, y):
  print("CONVERGENT")
else:
  print("NOT CONVERGENT")
```

## Least Square Regression
https://pythonnumericalmethods.berkeley.edu/notebooks/chapter16.04-Least-Squares-Regression-in-Python.html


```python
import numpy as np

x = [3,1,7,3,1,8,10,3,9,11,3,7,8,1,4,2]
y = [2,5,2,3,6,8,12,7,5,10,6,7,9,3,5,5]

x = np.array(x)
y = np.array(y)

A = np.vstack([x, np.ones(len(x))]).T
# print(A)

Y = y[:, np.newaxis]
# newaxis -> buat jadi array sendiri-sendiri
# print(Y)

# direct inverse
# alpha = np.dot(np.dot(np.linalg.inv(np.dot(A.T, A)), A.T), Y)

# pseudo inverse
pinv = np.linalg.pinv(A)
alpha = np.dot(pinv, Y) #pinv.dot(Y)

# print(alpha)

m = alpha[0]
b = alpha[1]

y_garis = m*x + b

# plotting
import matplotlib.pyplot as plt

plt.plot(x, y, '.', label="x y coordinates")
plt.plot(x, y_garis, 'r', label="least square regression")
plt.title("Least Square Regression")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Linear Interpolation

```python
from scipy.interpolate import interp1d

x1 = np.array([1, 6, 11, 16])
y1 = np.array([2, 5, 3, 7])

f = interp1d(x1, y1)
y_new = f(1.5)

# plotting
plt.plot(x1, y1, 'b', label="x y coordinates")
plt.plot(1.5, y_new, 'ro', label="linear interpolation at x = 1.5")
plt.title('Linear Interpolation at x = 1.5')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Cubic Spline Interpolation

```python
from scipy.interpolate import CubicSpline

x2 = [0, 1, 2]
y2 = [1, 3, 2]

x2 = np.array(x2)
y2 = np.array(y2)

f_cubic = CubicSpline(x2, y2)
x_new = np.linspace(0, 2, 100)
y_new = f_cubic(x_new)

# plot
plt.plot(x_new, y_new, 'b', label="cubic spline interpolation 0<=x<=2")
plt.plot(x2, y2, 'ro', label="x y coordinates")
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Root of Equation Initialisation

```python
import numpy as np

def f(x):
    return x**2-2

def g(x):
    return x*2
```

## Bisection Method

```python
def bissection(a, b, tol = 0.1):
    if(np.sign(f(a)) == np.sign(f(b))):
        print("Invalid Values")
        return
    
    m = (a + b) / 2

    if(np.abs(f(m)) < tol):
        print("Root: ", m)
        return
    elif(np.sign(f(a)) == np.sign(f(m))):
        return bissection(m,b,tol)
    elif(np.sign(f(b)) == np.sign(f(m))):
        return bissection(a,m,tol)
```

## Newton-Rhapson Method

```python
def newton_rapshon(x0, tol = 0.1):
    if(np.abs(f(x0)) < tol):
        print("Root: ", x0)
        return

    x1 = x0 - f(x0)/g(x0)
    return newton_rapshon(x1,tol)
```

```python
bissection(0,2,0.01)
newton_rapshon(1.5, 1e-6)
```

## Riemanns Integral

```python
import numpy as np
import scipy.integrate as integrate

#Pembatas
a = 0
b = 8
n = 5 #n=garis; kotak = n-1 = 4
h = (b-a)/(n-1)

x = np.linspace(a, b, n)

#Persamaan
f = lambda x: x**2

# def f(x):
#     return x**2

#left riemann
x_LR = x[:n-1]
I_LR = h * np.sum(f(x_LR)) #masing masing di sum terus dikali
print(f"left rieman : {I_LR}")

#right riemann
x_RR = x[1:n]
I_RR = h * np.sum(f(x_RR))
print(f"right rieman : {I_RR}")

#mid riemann
# x_MR = (x[:n-1] + x[1:n]) / 2
x_MR = (x_LR + x_RR) / 2
I_MR = h * np.sum(f(x_MR))
print(f"mid rieman : {I_MR}")
```

## Trapezoid Rule

```python
#cara 1
x_Trapezoid = np.hstack((x[0], x[1:n-1], x[1:n-1], x[n-1]))
I_Trapezoid1 = h/2 * np.sum(f(x_Trapezoid))
print(f"trapezoid1 : {I_Trapezoid1}")

#cara 2
y = f(x)
I_Trapezoid2 = h/2 * (y[0] + 2*np.sum(y[1:n-1]) + y[n-1])
print(f"trapezoid2 : {I_Trapezoid2}")
```
